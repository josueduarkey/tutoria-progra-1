### Tarea 5. Sistema de Gestión de una Tienda Escolar
Diseña un programa completo que permita realizar las siguientes funciones:

- Menú Principal: Mostrar opciones (1. Agregar producto, 2. Ver carrito, 3. Calcular total, 4. Salir).
- Agregar: Solicitar nombre del producto y su precio. Guardar los nombres en una lista y los precios en otra.
- Ver Carrito: Usar un for para mostrar cada producto con su respectivo precio lado a lado.
- Calcular Total: Sumar todos los precios. Si el total supera los $50.00, aplicar un descuento del 10% (usando if).
- Ciclo: El programa debe repetirse hasta que el usuario elija la opción 4.

```python


def addProductAndPrice(list_products, list_price):
    product = input('Digite el producto: ')
    list_products.append(product)
    price = float(input(f'Digite el precio de {product}: '))
    list_price.append(price)

    
def watchProducts(list_products, list_price):
     for prod, price in zip(list_products, list_price):
         print(f'{prod}: {price}')

def totalPagar(list_price):
    total = 0
    for price in list_price:
        total += price 
    return total

lista = []
prices = []

while (True):
    print('==========================')
    print('Digite la opción: ')
    print('1) Agregar producto')
    print('2) Ver carrito')
    print('3) Calcular total')
    print('4) Salir')
    print('==========================')

    
    option = int(input('Digite la opción que desee realizar (1-4): '))

    match option:
        case 1:
            addProductAndPrice(lista, prices)
        case 2:
            watchProducts(lista, prices)
        case 3:
            total = totalPagar(prices)
            print(f'El total es: {total}')
        case 4:
            print('Vuelva pronto')
            break
        case _:
            print('Digite una opción valida')
            continue

```
