## Bloque 2: Ciclos y Bucles (For y While)

### Tarea 1. Registro de Temperaturas
Crea un programa que solicite al usuario ingresar 7 lecturas de temperatura (una para cada día de la semana). 
Al final, el programa debe mostrar:

1. El promedio de las temperaturas **(usar un acumulador)**.
2. Cuántos días hubo temperaturas bajo cero **(usar un contador)**.

```python
cont = 0
suma = 0
while(cont < 7):
    temperatura = int(input(f'{cont} - Dato'))
    suma += temperatura
    cont += 1
print(f'El promedio es {suma/7}')
```

### Tarea 5. Control de Inventario 
Un almacén necesita registrar productos hasta que el usuario decida detenerse ingresando la palabra "salir". 
- Por cada producto, se pide el precio.
- El programa debe mostrar la suma total de la compra al final.

```python
total_pagar = 0
while(True):
    product = input('Digite el nombre del producto')
    if product.lower() == 'salir':
        print('ADIOOS')
        break
    precio = float(input('Digite un precio: '))
    total_pagar = precio + total_pagar
    print(f'Hola soy la var: {total_pagar}')

print(total_pagar)
```
