## Bloque 1: Decisiones Lógicas (Estructura if-elif-else)

### Tarea 1. El Validador de Voltaje
Crea un programa que solicite al usuario el voltaje detectado en un sensor industrial.

- Si el voltaje es menor a 100, mostrar "Bajo Voltaje".
- Si está entre 100 y 220 (inclusive), mostrar "Voltaje Estable".
- Si es mayor a 220, mostrar "Alerta: Sobrecarga".
- Si el usuario ingresa un número negativo, mostrar "Error: Lectura inválida".

```python
voltaje = int(input('Digite el voltaje: '))

if voltaje < 100:
    print('Bajo Voltaje')
elif voltaje >= 100 and voltaje <= 220:
    print('voltaje Estable')
elif voltaje > 220:
    print('Alto')
else :
    print('Error: Lectura invalida')
```

### Tarea 3. Sistema de Calificaciones Académicas

Un estudiante ingresa su nota final (0-10) y su porcentaje de asistencia (0-100). 
El programa debe decidir el estado del alumno:

- "Aprobado": Si la nota es $\ge 6$ y la asistencia es $\ge 80\%$.
- "Reprobado por Asistencia": Si la nota es $\ge 6$ pero la asistencia es $< 80\%$.
- "Reprobado por Nota": Si la nota es $< 6$.
- "Exento de Examen Final": Si la nota es $\ge 9$ y la asistencia es $100\%$.
