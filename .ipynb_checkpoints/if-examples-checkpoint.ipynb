## Bloque 1: Decisiones Lógicas (Estructura if-elif-else)

### Tarea 1. El Validador de Voltaje
Crea un programa que solicite al usuario el voltaje detectado en un sensor industrial.

- Si el voltaje es menor a 100, mostrar "Bajo Voltaje".
- Si está entre 100 y 220 (inclusive), mostrar "Voltaje Estable".
- Si es mayor a 220, mostrar "Alerta: Sobrecarga".
- Si el usuario ingresa un número negativo, mostrar "Error: Lectura inválida".

### Tarea 2. Clasificador de Triángulos por Ángulos
Solicita al usuario los tres ángulos internos de un triángulo (en grados). 

- Primero, valida si la suma es exactamente $180^\circ$. Si es válido, determina si el triángulo es:
- Rectángulo: Si tiene un ángulo de $90^\circ$.
- Obtusángulo: Si tiene un ángulo mayor a $90^\circ$.
- Acutángulo: Si todos sus ángulos son menores a $90^\circ$.
